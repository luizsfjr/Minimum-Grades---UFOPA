# Pegando links dos PSR


```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
root_url = 'https://ufopa.edu.br/proen/processos-seletivos/regular/psr-unificado-2023/'
response = requests.get(root_url)
content = response.content

soup = BeautifulSoup(content,'html.parser')

todos_li = soup.find('div',id='conteudo-paginas').find_all('li')

dados = {
    'Link':[],
    'Curso':[],
    'Chamada':[]
}

count = 0
chamada = 0
for i in todos_li:

  if(i.text.strip() == 'Bacharelado em Direito'):
    count += 1

  if(count == 1):
    chamada = 2
  elif(count == 2):
    chamada = 1

  if(i.text.startswith('L') or i.text.startswith('B')):

    dados['Link'].append(i.a.get('href'))
    dados['Curso'].append(i.text)
    dados['Chamada'].append(chamada)

df_2022 = pd.DataFrame(dados)
df_2022.to_csv('PSR_2023_Links_edicao.csv',index=False)
```

## Função de checagem de integridade

```python
def check_consistence(text,course):

  import re

  if any(c in text for c in ('óbidos', 'juruti', 'oriximiná')):
    curso, cidade = course.split('(')[0],course.split('(')[1]
    regex = f'.*{curso}.* - {cidade.replace(")","")}.*'
    #print(text,regex)
    if(re.match(regex,text)):
      return True
    return False
  else:
    if(text.find(course) != -1):
      return True
    return False
```

```python
import io

import requests
from PyPDF2 import PdfReader
import warnings
warnings.filterwarnings('ignore')
```

### Função que extrai notas

```python
def extract_scores(contents,modalidade):
  for linha in contents:
    index_ac = linha.lower().find(modalidade.lower())
    if(index_ac != -1):
      return linha.split(':')[1][:-1].strip()
  return None
```

```python
df_2020 = pd.read_csv('PSR_2023_Links_edicao.csv')
```

```python
df = df_2020

dados = {
    'Curso':[],
    'Chamada':[],
    'Ampla Concorrência':[],
    'Grupo 1':[],
    'Grupo 2':[],
    'Grupo 3':[],
    'Grupo 4':[],
    'Grupo 5':[],
    'Grupo 6':[],
    'Grupo 7':[],
    'Grupo 8':[],
}

with open('log.txt','w') as log:

  for link, curso, chamada in zip(df['Link'],df['Curso'], df['Chamada']):

    dados['Curso'].append(curso)
    dados['Chamada'].append(chamada)

    r = requests.get(link)
    f = io.BytesIO(r.content)
    reader = PdfReader(f)

    # Junta todas as páginas
    contents = []
    for page in range(len(reader.pages)):
      contents.extend(reader.pages[page].extract_text().split('\n'))

    # Extrai
    for modalidade in list(dados.keys())[2:]:
        try:
          dados[modalidade].append(extract_scores(contents,modalidade))
        except:
          dados[modalidade].append(None)
          log.write(f'ERRO no ***** {curso} - {chamada} - {modalidade} *****\n')

dados_finais = pd.DataFrame(dados)
```
